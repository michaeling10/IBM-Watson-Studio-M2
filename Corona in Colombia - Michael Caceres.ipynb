# Covid-19 cases in Colombia

Below we observe the behavior of the number of cases of the virus in Colombia, for this we will carry out the following steps:

    1. Import Libraries
    2. Define country and columns
    3. Illustrate the graph

# 1. Import Libraries

```python
import pandas as pd
import seaborn as sns

url = 'https://www.ecdc.europa.eu/sites/default/files/documents/COVID-19-geographic-disbtribution-worldwide-2020-03-26.xlsx'
df_raw = pd.read_excel(url)

df_raw['Countries and territories'].unique()
```

# 2. Define Country

```python
df = df_raw

df = df[df['Countries and territories']=='Colombia']

df = df.sort_values(['Year', 'Month' ,'Day'], ascending=[1, 2, 15])

df['ts'] = pd.to_datetime(df[['Year', 'Month' ,'Day']])
```

# 3. Graph


```python
sns.lineplot(x='ts', y='Cases', data=df, hue='Countries and territories')
```
